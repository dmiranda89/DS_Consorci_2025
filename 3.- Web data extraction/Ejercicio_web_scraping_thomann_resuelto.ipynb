## Ejemplo de Web Scraping del marketplace thomann


 - Busca en la página web https://www.thomann.de/es/cat.html y obtén un listado de todos los artículos que están a la venta incluyendo la mayor cantidad posible de información que consigas (nombre producto, precio, descripción, componentes...)

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options


```

#### Para usuarios que ejecuten en google colab

```python
# Utilizando Firefox como webdriver
from selenium.webdriver.firefox.options import Options as Options_firefox
# Instalar Firefox y geckodriver
!apt-get update
!apt-get install -y firefox-geckodriver firefox
# Configurar el WebDriver de Firefox
options = Options_firefox()
options.add_argument('--headless') # Modo headless para Colab
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
# Inicializar el driver de Firefox
driver = webdriver.Firefox(options=options)
# Iniciar el navegador Google Chrome
driver = webdriver.Chrome()
```

```python
# URL principal de Carrefour Supermercado
url = "https://www.thomann.de/es/cat.html"
# url = "https://www.thomann.de/es/index.html"
driver.get(url)

# Esperar a que la página cargue completamente
time.sleep(5)

# Extraer todos los elementos <a> (enlaces) de la página
links_elements = driver.find_elements(By.TAG_NAME, "a")

# Lista para guardar los enlaces únicos
links = set()

# Recorrer los elementos y obtener el texto y el href
for link in links_elements:
    try:
        url = link.get_attribute("href")
        if url:  # Solo guardar si el enlace no está vacío
            links.add(url)
    except:
        continue

# Cerrar el navegador
driver.quit()
```

```python
# links
# guitarras_y_bajos
# drums_y_percusion
# instrumentos_de_teclas
# equipo_de_estudio_y_recording
# software
clases_articulos = ['guitarras','bajo','drum','teclado','amplificad','piano','organo','microf','software']
# clases_articulos = ['amplificad']
articulos = []
for clases in clases_articulos:
    for link in sorted(links):  # Ordenar alfabéticamente
        if clases in link:
            
            articulos.append(link)
articulos = set(articulos)
print(len(articulos))
```

```python
response = requests.get('https://www.thomann.de/es/e-drums.html')

# Verificar que la solicitud fue exitosa
if response.status_code == 200:
    # Analizar el contenido HTML de la página
    soup = BeautifulSoup(response.text, "html.parser")
print(soup)
# # Buscar todos los productos en la página
# for item in soup.find_all("div", class_="fx-carousel__item"):
```

```python
# Encabezados para la solicitud HTTP
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
}
df = pd.DataFrame([])
for url in articulos:
#     print(url)
    # Realizar la solicitud HTTP
    response = requests.get(url, headers=headers)

    # Verificar que la solicitud fue exitosa
    if response.status_code == 200:
        # Analizar el contenido HTML de la página
        soup = BeautifulSoup(response.text, "html.parser")

    # Buscar todos los productos en la página
    for item in soup.find_all("div", class_="fx-carousel__item"):
        # Extraer el nombre del producto
        nombre_tag = item.find("div", class_="description fx-text")
        nombre = nombre_tag.get_text(strip=True) if nombre_tag else "N/A"

        # Extraer el precio del producto
        precio_tag = item.find("span", class_="price__primary")
        precio = precio_tag.get_text(strip=True) if precio_tag else "N/A"

        # Agregar los datos a la lista
        productos.append({"Nombre": nombre, "Precio": precio})
#         print(len(productos))
    # Crear un DataFrame con los datos
    df0 = pd.DataFrame(productos)
    df0['categoria'] = url.split('/')[-1].split('.html')[0].replace('_', ' ')
    df = pd.concat([df, df0], axis = 0)
df.shape
```
